```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# One-hot encode categorical features (e.g., Symptoms, Treatment Received, Condition)
def preprocess_data(df):
    categorical_columns = ['Symptoms', 'Treatment Received', 'Condition']
    encoder = OneHotEncoder(sparse_output=False)  # Use sparse_output instead of sparse
    encoded_data = encoder.fit_transform(df[categorical_columns])

    # Combine the encoded data with numerical columns (Age)
    numerical_columns = df[['Age']].values
    processed_data = np.hstack((numerical_columns, encoded_data))

    return processed_data

# Function to calculate similarity score
def calculate_similarity(patient_id, data):
    processed_data = preprocess_data(data)

    # Get data for the specific patient
    patient_data = data[data['Patient ID'] == patient_id]
    if patient_data.empty:
        raise ValueError("Patient ID not found in dataset.")

    # Get data of the patient's cluster and calculate similarity
    patient_index = data.index[data['Patient ID'] == patient_id].tolist()[0]
    similarity_matrix = cosine_similarity(processed_data)

    return similarity_matrix[patient_index]

# Example usage
similarity_scores = calculate_similarity(patient_id=1, data=df)  # Replace with actual patient ID
print(similarity_scores)
```
